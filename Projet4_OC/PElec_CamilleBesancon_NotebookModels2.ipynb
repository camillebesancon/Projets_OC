```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import scipy.stats
import sklearn.metrics as metrics
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import cross_validate, cross_val_score, RandomizedSearchCV
```

# Chargement des données et encodage

```python
df_ml=pd.read_csv('DataML.csv', sep=',')
df_ml.columns.tolist()
df_ml
```

```python
#Correlations des var. finales par rapport aux targets
corel = df_ml.corr().round(decimals=2)
fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(corel[['SiteEUI(kBtu/sf)','GHGEmissionsIntensity']], annot=True)
```

```python
enc = OneHotEncoder(handle_unknown='ignore')
# Objet OneHotEncoder pour données catégorielles 

enc_df = pd.DataFrame(  enc.fit_transform(df_ml[['LargestPropertyUseType', 'SecondLargestPropertyUseType', 'BuildingType']]).toarray() )
# fit_transform sur les données à encoder, qui sont ensuite re-converties en dataframe

df_ml= df_ml.join(enc_df)
df_ml.drop(columns=['LargestPropertyUseType', 'SecondLargestPropertyUseType', 'BuildingType'],inplace=True)
# Jointure sur les données et suppression des anciennes colonnes (remplacées par les versions encodées)

df_ml
```

### Variable noscore :
Booléen pour configurer les noms de fichiers / les données en input des modèles afin de comparer la performance des modèles en fonction de l'ajout ou non de l'EnergyStarScore (ESC) dans les données.
- noscore = False : données sans ESC
- noscore = True : Données avec ESC

```python
noscore=True
```

# Second modèle - Emissions

```python
MLliste=df_ml.columns.tolist()

#Initialisation en fonction de la valeur de "noscore"

if noscore:
    MLliste.remove('ENERGYSTARScore')
    name='NotarScore_'
else:
    name='WithStarScore_'
```

```python
#Création du répertoire pour écriture des outputs
rep='modele2_Emissions_'+name
if not rep in os.listdir('.'): 
    os.mkdir(rep)

#Pour écriture des métriqsues
metrik=open('Metrics_'+name+'.csv', "w")
print(MLliste)
```

```python
# Sélection des variables
selection=df_ml[MLliste]
X2=RobustScaler().fit_transform(selection)
X2=pd.DataFrame(X2,columns=selection.columns)
y2=X2['GHGEmissionsIntensity']
X2=X2.drop(columns=['SiteEUI(kBtu/sf)','GHGEmissionsIntensity'])
y2
```

```python
# Cross validation - stratified k-fold avec 10 splits et 10 repeats. On a 4860 "individus", donc 486 par split
# Jeu de d'entrainement et test, split 80/20 - 972 individus pour le test
cv=RepeatedKFold(n_splits=10,n_repeats=10)
```

```python
# Split des données
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.20)

# Creation du régresseur, choix du randomforest car c'est une bonne solution générale
rf = RandomForestRegressor(n_estimators = 500, random_state = 42)

# Grille de paramètres
random_grid = {
    'bootstrap': [True, False],
    'max_depth': [3, 10, None],
    'min_samples_leaf': [1, 2, 3],
    'max_features': [1, 2],
    'n_estimators': [100, 300, 500]}

#Cross validation et sélection des hyperparamètres avec la même grille que pour le 1er modèle
clf2 = RandomizedSearchCV(rf, random_grid, random_state=0,cv=cv)
search2 = clf2.fit(X_train2, y_train2)
```

```python
search2.get_params()
```

```python
#Scores

score2=search2.score(X_train2, y_train2)
print("R-squared train:", score2)

y_pred2=search2.predict(X_test2)
y_pred2

score22=search2.score(X_test2, y_test2)
print("R-squared test:", score22)
```

```python
# Métriques 

mse2 = metrics.mean_squared_error(y_test2, y_pred2)
mae2= metrics.mean_absolute_error(y_test2, y_pred2)
print("MSE: ", mse2)
print("RMSE: ", mse2*(1/2.0)) # root mse (rmsd toussa, même principe)
print("MAE: ", mae2)


metrik.write(',R-squared train,R-squared train,MSE,RMSE,MAE\nEmissions GES'+str(score2)+','+str(score22)+','+str(mse2)
              +','+str(mse2*(1/2.0))+','+str(mae2)+'\n')

metrik.close()
```

```python
# Visualisation des données prédites / réelles 
figg, axx = plt.subplots(figsize=(15,11))
x_axx = range(len(y_test2))
x_axx2 = range(len(y_test2))
x_axx2 = np.array(x_axx2)
plt.plot(x_axx, y_test2, linewidth=1.1, label="original")
plt.plot(x_axx2+2, y_pred2, linewidth=1.1, label="predicted")
plt.title("y-test and y-predicted data")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='best',fancybox=True, shadow=True, fontsize=14)
plt.savefig(rep+'/model2_CompPred_'+name+'.png')
plt.grid(True)
plt.show()
```

```python
# Feature importance + barplot
imp2=search2.best_estimator_.feature_importances_
order2=np.argsort(search2.best_estimator_.feature_importances_)[::-1]
imp2[order2][0:3]
```

```python
plt.title('Feature Importance')
plt.bar(range(0,15), imp2[order2][0:15], align='center')
plt.xticks(range(0,15), X_train2.columns[order2][0:15], rotation=90)
plt.tight_layout()
plt.savefig(rep+'/featureImp_model2'+name+'.png')
plt.show()
plt.close()
```
